```python
import pandas as pd 
```

```python
pew = pd.read_csv("pew.csv")
pew 
```

```python
pew = pew.rename(columns={
    'religion':'Religion',
    '<$10k': 'less than 10k',
    '$10-20k':'10-20k',
    '$20-30k':'20-30k',
    '$30-40k':'30-40k',
    '$40-50k':'40-50k',
    '$50-75k':'50-75k'
})  
pew.head(4) 
```

```python
pew 
```

```python
Pew = pd.melt(pew, id_vars =['Religion'],
             value_vars =['less than 10k', '10-20k', '20-30k', '30-40k', '40-50k', '50-75k'],
             value_name ='Frequency',
             var_name ='Income_Range($)')
Pew 
```
