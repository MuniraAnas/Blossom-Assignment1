```python
import pandas as pd 
```

```python
billboard = pd.read_csv('billboard.csv')
billboard.head(5) 
```

```python
billboard.dtypes 
```

```python
billboard['date.peaked'] = pd.to_datetime(billboard['date.peaked'])
billboard['date.entered'] = pd.to_datetime(billboard['date.entered'])
billboard.head(3) 
```

```python
billboard.dtypes 
```

```python
Billboard_melted = pd.melt(billboard, id_vars =['artiste', 'track', 'genre', 'date.entered', 'date.peaked']) 
Billboard_melted 
```

```python
Billboard_melted = Billboard_melted.rename(columns ={
    'variable' : 'week',
    'value' : 'rank'
})
Billboard_melted 
```

```python
Billboard_melted['week'] = Billboard_melted['week'].str.extract('(\d+)').astype(int)

Billboard_melted.head(6) 
```

```python
Billboard_melted['week'] = Billboard_melted.week.astype(str) 

Billboard_melted.dtypes 
```

```python
Billboard_melted['week'] = Billboard_melted['week'].str.extract('(\d+)').astype(int)

Billboard_melted.head(5) 
```

```python
Billboard_melted['entered_to_peak'] = Billboard_melted['date.peaked']- Billboard_melted['date.entered']

Billboard_melted.head(10) 
```

```python
Billboard_melted['entered_to_peak'] = Billboard_melted['entered_to_peak'].dt.days 

Billboard_melted.head(5) 
```

```python
Billboard_melted.dtypes 
```

```python
Billboard_melted 
```

```python
Billboard_ans = Billboard_melted[['artiste', 'track', 'rank', 'genre']][Billboard_melted['entered_to_peak'] == 35]
Billboard_ans 
```

```python
Billboard_ans[['artiste','track','rank','genre']][Billboard_melted['genre']=='Rock'].sort_values('rank').head(15)  
```
